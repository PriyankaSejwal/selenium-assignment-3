Q2. In first ques, from the first three page scrape the "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”.

```python
#extracting urls of three pages
url=driver.find_elements_by_xpath("//ul[@class='a-pagination']//li/a")
urls=[i.get_attribute('href') for i in url[0:3]]
urls
```

```python
#extracting url of each product
link=[]
for i in urls:
    driver.get(i)
    for j in driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']"):
        link.append(j.get_attribute('href'))
link
```

```python
len(link)
```

```python
#scraping information
brand=[]
prod=[]
price=[]
exch=[]
Del=[]
avail=[]
url=[]
rating=[]
no=[]
details=[]
for i in link:
    driver.get(i)
    try:
        brand.append(driver.find_element_by_xpath("//div[@class='a-section a-spacing-none']/a").text)
    except NoSuchElementException as e:
        brand.append("--")
        
#product
    prod.append(driver.find_element_by_id('title').text)

    
#price
    try:
        price.append(driver.find_element_by_xpath("//td[@class='a-span12']/span").text)
    except NoSuchElementException:
        price.append("--")

#return/exchange
    try:
        exch.append(driver.find_element_by_xpath("//div[@id='RETURNS_POLICY']").text)
    except NoSuchElementException as e:
        exch.append("--")

#delivery
    try:
        Del.append(driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b").text)
    except NoSuchElementException as e:
        Del.append("--")
        
#availability
    try:
        avail.append(driver.find_element_by_xpath("//div[@id='availability']/span").text)
    except NoSuchElementException as e:
        avail.append("--")
        
#prod url
    url.append(i)

#rating
    try:
        rating.append(driver.find_element_by_xpath("//span[@data-hook='rating-out-of-text']").text)
    except NoSuchElementException as e:
        rating.append("--")
        
#no.of ratings
    try:
        no.append(driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']").text)
    except NoSuchElementException as e:
        no.append("--")
        
#other details
    try:
        details.append(driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']").text)
    except NoSuchElementException:
        details.append("--")
```

```python
df=pd.DataFrame({'Brand':brand,'Name of product':prod,'Rating':rating,'No.of.Ratings':no,'Price':price,'Return/exchange':exch,'Expected Delivery':Del,'Availability':avail,'Other details':details,'Product url':url})
df
```

```python
#clicking on the desired element- top gaming laptops 
driver.find_element_by_xpath("//div[@class='listing_container']//li[9]/a").click()
```

```python
#scraping the information of the laptops in empty lists
Name=[]
price=[]
seller=[]
OS=[]
display=[]
processor=[]
memory=[]
weight=[]
dim=[]
gp=[]

    

#PRODUCT NAME
for j in driver.find_elements_by_tag_name('h3'):
        Name.append(j.text)

#PRICE
for i in driver.find_elements_by_xpath("//td[@class='smprice']"):
    price.append(i.text)
    
#SELLER
for i in driver.find_elements_by_xpath("//td[@class='smmerchant']"):
    seller.append(i.text)
    
#OS
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[3]/td[3]"):
    OS.append(i.text)
    
#DISPLAY
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[4]/td[3]"):
    display.append(i.text)
    
#PROCESSOR
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[5]/td[3]"):
    processor.append(i.text)
    
#MEMORY
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[6]/td[3]"):
    memory.append(i.text)
    
#WEIGHT
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[7]/td[3]"):
    weight.append(i.text)
    
#DIMENSION
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[8]/td[3]"):
    dim.append(i.text)
    
#GRAPHICS PROCESSOR
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[9]/td[3]"):
    gp.append(i.text)
```

```python
#extracting hrefs of the laptops to scrape more information
url=[]
for i in driver.find_elements_by_xpath("//div[@class='TopNumbeHeading sticky-footer' or @class='TopNumbeHeading active sticky-footer' or @class='TopNumbeHeading sticky-footer active']//a"):
    url.append(i.get_attribute('href'))
url
```

```python
status=[]
website=[]
launch=[]
dig_rate=[]
pros=[]
cons=[]
specs=[]
link=[]
for i in url:
    driver.get(i)
#MARKET STATUS
    try:
        status.append(driver.find_element_by_xpath("//div[@class='Block-status']/b").text)
    except NoSuchElementException as e:
        status.append("--")
        
#OFFICIAL WEBSITE
    try:
        website.append(driver.find_element_by_xpath("//div[@class='status-bar only-desktop']//div[3]/b").text)
    except NoSuchElementException as e:
        website.append('--')

#LAUNCH DATE
    try:
        launch.append(driver.find_element_by_xpath("//div[@class='status-bar only-desktop']//div[2]/b").text)
    except NoSuchElementException as e:
        launch.append('--')

#DIGIT RATING
    try:
        dig_rate.append(driver.find_element_by_xpath("//div[@class='Numeric-Number greentext' or @class='Numeric-Number redtext' or @class='Numeric-Number yellowtext']").text)
    except NoSuchElementException as e:
        dig_rate.append("--")


#PROS
    try:
        pros.append(driver.find_element_by_xpath("//div[@class='pros-Cons']//ul[1]").text.strip())
    except NoSuchElementException as e:
        pros.append("--")
        
#CONS
    try:
        cons.append(driver.find_element_by_xpath("//div[@class='pros-Cons']//ul[2]").text.strip())
    except NoSuchElementException as e:
        cons.append("--")
        
#KEY SPECIFICATIONS
    try:
        specs.append(driver.find_element_by_xpath("//div[@class='Key-specs']").text.strip())
    except NoSuchElementException as e:
        specs.append("--")
        
#LAPTOP LINK
    link.append(i)
```

```python
laptop=pd.DataFrame({'Laptop Name':Name,
'Price':price,
'Seller':seller,
'Operating System':OS,
'Display':display,
'Processor':processor,
'Memory':memory,
'Weight':weight,
'Graphics Processor':gp,
'Dimensions':dim,
'Market Status':status,
'Official Website':website,
'Launch Date':launch,
'Digit Rating':dig_rate,
'Pros':pros,
'Cons':cons,
'Key Specifications':specs,
'Laptop URL':link})


laptop
```

```python
laptop.to_csv('top_gaming_laptops.csv')
```
